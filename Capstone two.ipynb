```python
def scatterplots(columns, ncol=None, figsize=(15, 8)):
    if ncol is None:
        ncol = len(columns)
    nrow = int(np.ceil(len(columns) / ncol))
    fig, axes = plt.subplots(nrow, ncol, figsize=figsize, squeeze=False)
    fig.subplots_adjust(wspace=0.5, hspace=0.6)
    for i, col in enumerate(columns):
        ax = axes.flatten()[i]
        ax.scatter(x = col, y = "mobile_price", data=df, alpha=0.5)
        ax.set(xlabel=col, ylabel='Mobile price')
    nsubplots = nrow * ncol    
    for empty in range(i+1, nsubplots):
        axes.flatten()[empty].set_visible(False)
```

```python
features = [columns for columns in df.columns if columns not in ["mobile_name", "mobile_price","dual_sim","num_cores"]]
```

```python
scatterplots(features, ncol=3, figsize=(15, 15))
```
